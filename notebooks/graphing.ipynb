```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"...\testing_results_.xlsx" #change to your directory
df = pd.read_excel(file_path)

df["best loss"] = pd.to_numeric(df["best loss"], errors="coerce")
df["Earth Mover's Distance"] = pd.to_numeric(df["Earth Mover's Distance"], errors="coerce")
```

```python
plt.figure(figsize=(27, 5))
plt.plot(df.index, df["best loss"], linestyle='-', linewidth=0.5, color='purple', markersize=5)

plt.xlabel("Value", fontsize=12)
plt.ylabel("Best Loss", fontsize=12)
plt.grid(alpha=0.3)  
plt.legend(["Best Loss"], fontsize=12)

plt.savefig("best_loss_plot.pdf", format="pdf", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(27, 5))
plt.plot(df.index, df["Earth Mover's Distance"],  linestyle='-', linewidth=0.5, color='orange', markersize=5)

plt.xlabel("Index")
plt.ylabel("Value")
plt.legend(["Earth Mover's Distance"], fontsize=12)
plt.grid()

plt.savefig("Earth Mover's Distance.pdf", format="pdf", dpi=300, bbox_inches="tight")

plt.show()
```
